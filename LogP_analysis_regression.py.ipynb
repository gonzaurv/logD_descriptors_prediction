# Lineal regression for all the possible combination in the Data set


```python
import time
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
import pandas as pd

def regresion_lineal_multiple(data_frame_X, Y):
    matriz_datos = []
    x1_para_borrar = []
    start = time.time()
    
    for a in range(len(data_frame_X.columns.values)):
        X1 = data_frame_X.iloc[:, a:a+1]
        x1_name = X1.columns.values
        x1_name_tolist = x1_name.tolist()
        reg1 = LinearRegression().fit(X1, Y)
        intercept1 = reg1.intercept_
        coeff1 = reg1.coef_
        preds1 = reg1.predict(X1)
        mae1 = mean_absolute_error(Y, preds1)
        mse1 = mean_squared_error(Y, preds1)
        rmse1 = np.sqrt(mse1)
        temporal_data1 = {'x1': x1_name, 'x2': '', 'Score': reg1.score(X1, Y),
                          'Intercept': intercept1[0], 'Coef_1': coeff1[0, 0], 'Coef_2': '',
                          'MAE': mae1, 'MSE': mse1, 'RMSE': rmse1}
        matriz_datos.append(temporal_data1)
        x1_para_borrar.append(x1_name)
        
        for b in range(len(data_frame_X.columns.values)):
            X2 = data_frame_X.iloc[:, b+1:b+2]
            x2_name = X2.columns.values
            x2_name_tolist = x2_name.tolist()
            df_test = pd.concat([X1, X2], axis=1, join='outer')
            
            if (df_test.shape[1] == 2) and (x1_name_tolist != x2_name_tolist):
                reg2 = LinearRegression().fit(df_test, Y)
                intercept = reg2.intercept_
                coeff = reg2.coef_
                score2 = reg2.score(df_test, Y)
                preds2 = reg2.predict(df_test)
                mae2 = mean_absolute_error(Y, preds2)
                mse2 = mean_squared_error(Y, preds2)
                rmse2 = np.sqrt(mse2)
                temporal_data = {'x1': x1_name_tolist, 'x2': x2_name_tolist, 'Score': score2,
                                 'Intercept': intercept[0], 'Coef_1': coeff[0, 0], 'Coef_2': coeff[0, 0] if coeff.shape[1] == 1 else coeff[0, 1],
                                 'MAE': mae2, 'MSE': mse2, 'RMSE': rmse2}
                if x2_name not in x1_para_borrar:
                    matriz_datos.append(temporal_data)
                    final_matrix = pd.DataFrame(matriz_datos)
    
    print(f'Tiempo total = ', round(time.time() - start, 3), 'segundos')
    return final_matrix


            
        
```

## **Correlation analysis**

The Variance Inflation Factor (VIF) is a parameter used to quantify multicollinearity in a linear regression model. Multicollinearity occurs when two or more predictor variables are highly correlated with each other. VIF is calculated for each predictor variable and provides a measure of how much the variance of a regression coefficient is inflated due to multicollinearity.

The formula for calculating the VIF of a specific predictor variable is:

\[ VIF_{\hat{\beta}^j} = \frac{1}{1 - R^2} \]

Where \( R^2 \) is obtained by fitting a linear regression model with the predictor variable \( X_j \) as the dependent variable and all other predictor variables as independent variables. A high \( R^2 \) indicates a strong correlation between the predictor variable \( X_j \) and the other predictor variables, resulting in a high VIF value.

Commonly used reference limits for interpreting VIF are:

- VIF = 1: Complete absence of multicollinearity.
- 1 < VIF < 5: Possible presence of multicollinearity that may moderately affect the regression model.
- 5 < VIF < 10: Presence of multicollinearity that may significantly affect the regression model.
- VIF > 10: High multicollinearity that can considerably distort the regression model results.

Tolerance, which is the inverse of VIF, provides an additional measure to assess multicollinearity. Recommended tolerance limits are between 1 and 0.1, where values closer to 1 indicate lower multicollinearity and values closer to 0.1 indicate higher multicollinearity.

In summary, VIF and tolerance are measures used to evaluate multicollinearity in a linear regression model, where higher VIF values and lower tolerance values indicate a greater presence of multicollinearity, which can affect the accuracy of the estimated coefficients in the model.
